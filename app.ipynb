```python
features = list(df_playersnew2.columns.values)
df_playersnew2 = df_playersnew2.fillna(0)
from sklearn.preprocessing import StandardScaler
# Separating out the features
x = df_players.loc[:, features].values
# Separating out the target
y = df_players.loc[:,['player']].values
# Standardizing the features
#x = StandardScaler().fit_transform(x)
x = np.nan_to_num(x)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['component1','component2'])
```

```python
df_players = df_players.reset_index(drop=True)
finalDf = pd.concat([principalDf, df_players[['player']]], axis = 1)
finalDf = pd.concat([finalDf, df_players[['squad']]], axis = 1)
```

```python
def find(player_name):
    player = player_name
    a = (finalDf[finalDf['player'] == player])['component1'].iloc[0]
    b = (finalDf[finalDf['player'] == player])['component2'].iloc[0]
    #print(a,b)
    finalDf['distance'] = (a-finalDf['component1'])*(a-finalDf['component1'])+(b-finalDf['component2'])*(b-finalDf['component2'])
    #finalDf['distance'] = (finalDf['component1'])
    print(((finalDf.sort_values(['distance'], ascending=[True])))[1:10])
```
